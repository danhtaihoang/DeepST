```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os, sys
```

```python
coef_mean = np.loadtxt("coef_heatmap_mean.txt")
coef_mean.shape
```

```python
coef0 = np.loadtxt("220Nov21BRCA_nonrot_ik4_il0/R.txt", dtype="str")
coef0 = coef0[:,1].astype(float)
coef0.shape
```

```python
coef1 = np.loadtxt("220Nov21BRCA_nonrot_ik4_il1/R.txt", dtype="str")
coef1 = coef1[:,1].astype(float)
coef1.shape
```

```python
coef01 = (coef0 + coef1)/2.
coef01.shape
```

```python
nx,ny = 3,1
fig, ax = plt.subplots(ny,nx,figsize=(nx*4,ny*3))

ax[0].plot(coef0, coef1, "o", markersize = "0.3")
ax[0].set_xlabel("Fold-1 correlations")
ax[0].set_ylabel("Fold-2 correlations")

ax[1].plot(coef01, coef_mean, "o", markersize = "0.3")
ax[1].set_xlabel("Correlations")
ax[1].set_ylabel("Model roburstness")

ax[2].hist(coef_mean,bins=10,histtype='bar',rwidth=0.8)
ax[2].set_xlabel("Model roburstness")
ax[2].set_ylabel("Number of genes")

plt.tight_layout(h_pad=1, w_pad= 1.5)
plt.savefig('roburstness.pdf', format='pdf', dpi=50)
```
